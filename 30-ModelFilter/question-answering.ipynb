```python
import pandas as pd
from transformers import pipeline
```

```python
model_name = "deepset/roberta-base-squad2"

# a) Get predictions
nlp = pipeline('question-answering', model=model_name, tokenizer=model_name, device=0)
QA_input = {
    'question': 'Why is model conversion important?',
    'context': 'The option to convert models between FARM and transformers gives freedom to the user and let people easily switch between frameworks.'
}
res = nlp(QA_input)
```

```python
comments = pd.read_csv('all_comments.csv', encoding_errors='ignore')
comments
```

```python
fun_labels = ['Dependable', 'Lasts long', 'Value for money', 'Quality fit and finish', 'Attractive styling'
                                    , 'Safe', 'Retains resale value', 'Driver comfort', 'Fun to drive', 
                                    'Advanced features', 'Responsive handling', 'Prestigious', 'Dealerships', 
        'Fuel efficient', 'Quick acceleration', 'Environmentally friendly', 'Affordable']
lower_labels = [i.lower() for i in fun_labels]
```

```python
QA_input = {
    'question': 'Is value for money mentioned?',
    'context': comments.at[0, 'Comment']
}
res = nlp(QA_input)
res
```

```python
comments[lower_labels] = pd.NaT
comments
```

```python
import time

start = time.time()

rown = comments.shape[0]

for i in range(rown):
  for j in lower_labels:
    try:
      QA_input = {
          'question': 'Is ' + j + ' mentioned?',
          'context': comments.at[i, 'Comment']
      }
      res = nlp(QA_input)
      comments.at[i, j] = 1 if res['score'] > 0.1 else 0 
    except:
      pass
    
end = time.time()
print((end - start)/60)

comments.to_csv('comments_exists.csv', index=False)

print("Save it")
```
