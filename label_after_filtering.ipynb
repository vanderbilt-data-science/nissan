```python
import pandas as pd
from transformers import pipeline
classifier = pipeline("zero-shot-classification",
                      model="facebook/bart-large-mnli", device=0)
```

```python
comments = pd.read_csv('comments_exists.csv', encoding_errors='ignore')
```

```python
sequence_to_classify = "I have to preface this by saying I purchased another new car recently and it ended up being a Lemon which the manufacturer repurchased (or is repurchasing because the process drags on forever). That said, I‚Äôm a little hyper sensitive to any issues with cars right now. I test drove the Rogue just to ‚Äúcheck it out‚Äù with no intention on purchasing so quickly after my Lemon experience but I fell in love with it. It‚Äôs roomy, has a ton of great features, crystal clear back up/surround camera, awesome moonroof, great MPG‚Äôs, etc etc.  Even though the features and look of the car is great, I noticed little glitches from day one. The cameras would switch on randomly when slowing the car during travel, the automatic breaking warning light would come on and then turn off, the radio was working and then sounded like I was in a tunnel and then resumed playing normally, AC blew hot air when on cool until I turned it down to ‚ÄúLo‚Äù, the touch screen acts crazy here and there, etc etc. For the most part I wrote these things off as ‚Äúuser error‚Äù or accepting that maybe the car just acted like this. BUT‚Ä¶today after owning the car for about two weeks/600 miles the emergency breaking malfunction light came on and remained on (I‚Äôm talking about the big, bright notice that pops up in your dash cluster). Like the other ‚Äúglitches‚Äù the malfunction light disappeared when I parked, turned off the car and then turned it on again. I realized that the glitches might be a little more than what I originally thought though. Having the malfunction light in addition to the warning symbols prompted me to take the car in to service. I‚Äôm currently waiting to find out what the issue is and how long it will take to fix. I also did some more review reading and I have to say that I‚Äôm really nervous that the car will have continuing issues. The reviews on this site are fairly positive but if you look around a bit more you find some scary stuff about this car. I really thought there was no possible way that I would have this type of experience with TWO brand new cars within a six month period but here we are. I hope this is just some weird full moon voodoo that gets repaired and never causes me an issue again but I‚Äôm really not that confident in that right now. All I‚Äôm saying is‚Ä¶be careful!"
candidate_labels = ['dependable', 'not dependable']
classifier(sequence_to_classify, candidate_labels, multi_label=False)
```

```python
dependable_labels = ['dependable', 'not dependable']
lasts_labels = ['lasts long', 'not lasts long']
value_labels = ['value for money', 'not value for money']
quality_labels = ['quality fit and finish', 'not quality fit and finish']
attractive_labels = ['attractive styling', 'not attractive styling']
safe_labels = ['safe', 'not safe']
retains_labels	= ['retains resale value', 'not retains resale value']
driver_labels	= ['driver comfort', 'driver not comfortable']
fun_labels = ['fun to drive', 'not fun to drive']
```

```python
# For first 9 columns
import time

start = time.time()

row_num = comments.shape[0]
for i in range(row_num):
  #for j in range(3,12):
  if comments.iat[i, 3] != 0:
      try:
        res_3 = classifier(comments.at[i, 'Comment'].lower(), dependable_labels, multi_label=False)
        comments.iat[i, 3] = res_3.get('scores')[0]*2-1
      except:
        pass

  if comments.iat[i, 4] != 0:
      try:
        res_4 = classifier(comments.at[i, 'Comment'].lower(), lasts_labels, multi_label=False)
        comments.iat[i, 4] = res_4.get('scores')[0]*2-1
      except:
        pass

  if comments.iat[i, 5] != 0:
      try:
        res_5 = classifier(comments.at[i, 'Comment'].lower(), value_labels, multi_label=False)
        comments.iat[i, 5] = res_5.get('scores')[0]*2-1
      except:
        pass

  if comments.iat[i, 6] != 0:
      try:
        res_6 = classifier(comments.at[i, 'Comment'].lower(), quality_labels, multi_label=False)
        comments.iat[i, 6] = res_6.get('scores')[0]*2-1
      except:
        pass

  if comments.iat[i, 7] != 0:
      try:
        res_7 = classifier(comments.at[i, 'Comment'].lower(), attractive_labels, multi_label=False)
        comments.iat[i, 7] = res_7.get('scores')[0]*2-1
      except:
        pass

  if comments.iat[i, 8] != 0:
      try:
        res_8 = classifier(comments.at[i, 'Comment'].lower(), safe_labels, multi_label=False)
        comments.iat[i, 8] = res_8.get('scores')[0]*2-1
      except:
        pass

  if comments.iat[i, 9] != 0:
      try:
        res_9 = classifier(comments.at[i, 'Comment'].lower(), retains_labels, multi_label=False)
        comments.iat[i, 9] = res_9.get('scores')[0]*2-1
      except:
        pass

  if comments.iat[i, 10] != 0:
      try:
        res_10 = classifier(comments.at[i, 'Comment'].lower(), driver_labels, multi_label=False)
        comments.iat[i, 10] = res_10.get('scores')[0]*2-1
      except:
        pass

  if comments.iat[i, 11] != 0:
      try:
        res_11 = classifier(comments.at[i, 'Comment'].lower(), fun_labels, multi_label=False)
        comments.iat[i, 11] = res_11.get('scores')[0]*2-1
      except:
        pass
    
end = time.time()
print((end - start)/60)

#comments.to_csv('comments_score.csv', index=False)

print("Save it")
```
