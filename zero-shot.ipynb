```python
import pandas as pd
from transformers import pipeline
classifier = pipeline("zero-shot-classification",
                      model="facebook/bart-large-mnli", device=0)
```

```python
sequence_to_classify = "one day I will see the world"
candidate_labels = ['This is about travel', 'This is about cooking', 'This is about world']
classifier(sequence_to_classify, candidate_labels)
```

```python
comments = pd.read_csv('all_comments.csv', encoding_errors='ignore')
```

```python
comments
```

```python
fun_labels = ['Dependable', 'Lasts long', 'Value for money', 'Quality fit and finish', 'Attractive styling'
                                    , 'Safe', 'Retains resale value', 'Driver comfort', 'Fun to drive', 
                                    'Advanced features', 'Responsive handling', 'Prestigious', 'Dealerships', 
        'Fuel efficient', 'Quick acceleration', 'Environmentally friendly', 'Affordable']
lower_labels = [i.lower() for i in fun_labels]
```

```python
comments[lower_labels] = pd.NaT
comments
```

```python
import time

start = time.time()

rown = comments.shape[0]
for i in range(rown):
    try:
      res = classifier(comments.at[i, 'Comment'].lower(), lower_labels, multi_label=True)
      comments.at[i, lower_labels] = dict(zip(res['labels'], res['scores']))
    except:
      pass
    
end = time.time()
print((end - start)/60)

comments.to_csv('comments_score.csv', index=False)

print("Save it")
```

```python
comments_score = pd.read_csv('comments_score.csv')
comments_score
```

```python
sum_scores = comments_score.groupby('Model')[lower_labels].sum()
```

```python
count_df = comments_score.groupby('Model')['Model'].count()
```

```python
sum_scores[lower_labels].div(count_df, axis=0)
```
